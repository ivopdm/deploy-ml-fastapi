```python
import pandas as pd
```

```python
census_data = pd.read_csv('data/census.csv')
```

```python
census_data.sample(10)
```

```python
census_data.info()
```

```python
# Function to transform object columns to category columns
def transform_to_category(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = df[col].astype('category')
    return df
```

```python
# dataframe shallow copy
census_data_cat = census_data.copy()
transform_to_category(census_data_cat)
```

```python
census_data_cat.info()
```

```python
census_data_cat.sample(10)
```

```python
census_data_cat.to_csv('data/census_cat.csv')
```
